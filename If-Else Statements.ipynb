### If-Else Statements

If-else statements are one of the most basic forms of logic used in modern day programming.

It literally translates to "computer, if this happens then do that. Otherwise if that happens, do this instead."

The program evaluates it in booleans. Meaning "if this is TRUE, then do this."

```python
if 5 > 4:
    print("Yes, 5 is greater than 4")
```

```python
are_you_hungry = True
if are_you_hungry == True:
    print("Go get some food!")
```

```python
if are_you_hungry:
    print("Go get some food!")
```

```python
my_age = 4
if my_age < 5:
    print("toddler")
    
elif my_age > 5 and my_age < 10:
    print("kiddo")

else:
    print("teenager? adult?")
```

```python
my_favorite_icecream = "strawberry"
if my_favorite_icecream == "vanilla":
    print("you guessed right")
else:
    print("you guessed wrong!")
```
